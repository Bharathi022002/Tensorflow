```python
#libraries
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Dense
from tensorflow.keras import Sequential
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
#reading the csv file
df=pd.read_csv('C:/Users/ELCOT/Downloads/breast_cancer.csv')
df
```

```python
#no use of id , so we are dropping it
df.drop(columns=['id'],inplace=True)  #inplace to remove permanently
```

```python
#after dropping id reading the csv file
df.head()
```

```python
#in x- droping diagnosis and storing all the colums except diagnosis
x=df.drop(columns='diagnosis')
#in y- droped diagnosis is stored
y=df['diagnosis']
```

```python
x
```

```python
y
```

```python
#checking whether all are numeric
x.info()
```

```python
#numerical variables can only be scaled
#if the dataset has non numerical value change to numeric using encoding method-one heart encoding
ss=StandardScaler()
ss.fit_transform(x)
#all are in array form
```

```python
# from above after scaling we got array value.. so changing array to normal numerical value
ss=StandardScaler()
x_scaled=pd.DataFrame(ss.fit_transform(x),columns=x.columns)
x_scaled
```

```python
#replacing m-0 ,b-1 
y=y.replace({'M':0,'B':1})
y
```

```python
#changing y to numeric, using one heart encoding
y_cat=keras.utils.to_categorical(y)
y_cat
```

```python
#training the dataset
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y_cat,test_size=0.2,random_state=8)
```

```python
#neural networks method
nn=Sequential()
nn.add(Dense(units=16,activation='relu',input_dim=30))   #hidden layer-1 #input_dim =no.of columns
nn.add(Dense(units=7,activation='relu')) #hidden layer-2
nn.add(Dense(units=3,activation='relu'))  #hidden layer-3
nn.add(Dense(units=2,activation='softmax')) #units-2  output 
nn.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy']) #metrices - to display all epochs accuracy
```

```python
# batch size - to split the columns based on batch size
#epochs - no of time it run
nn.fit(x_train,y_train,batch_size=30,epochs=100)
```

```python
nn.evaluate(x_test,y_test)
```

```python
nn.evaluate(x_train,y_train)
```
